```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
import seaborn as sns

import pandas as pd
df = pd.read_csv("Credit.csv")
df = df.iloc[:,1:7]
df
```

```python
df.plot(kind='scatter', x = 'Income', y = 'Limit', title = "Limit with Income")
sns.lmplot( x = 'Income', y='Limit', data = df)
```

```python
# Residual Income
r1 = sns.residplot('Income', 'Limit', data= df)
r1.set(xlabel = 'Income', ylabel = 'Residuals')
```

```python
# slope and the intercept for Income
# from the residual graph, since it is distributed evenly we can say that Income is a good predictor
lin_reg = LinearRegression()
lin_reg.fit(df[['Income']], df[['Limit']])
lin_reg.intercept_,lin_reg.coef_
```
