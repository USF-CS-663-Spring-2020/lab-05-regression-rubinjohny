```python
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
import seaborn as sns

import pandas as pd
df = pd.read_csv("Credit.csv")
df = df.iloc[:,1:7]
df
```

```python
df.plot(kind='scatter', x = 'Limit', y = 'Income', title = "Limit with Income")
sns.lmplot( x = 'Limit', y='Income', data = df)
```

```python
# Residual Income
r1 = sns.residplot('Limit', 'Income', data= df)
r1.set(xlabel = 'Limit', ylabel = 'Residuals')
```

```python
# slope and the intercept for Income
lin_reg = LinearRegression()
lin_reg.fit(df[['Income']], df[['Limit']])
lin_reg.intercept_,lin_reg.coef_
```
